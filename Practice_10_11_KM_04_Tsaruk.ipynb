Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
import re

while True:
    a = input("Enter the number: ")
    try:
        a = int(a)
        if a < 0:
            print("\033[31mincorrect number\033[37m")
            continue
    except ValueError:
        print("\033[31mincorrect number\033[37m")
    else:
        break
    

print("\033[32m")
list_start = [1]; n = 0
while len(list_start) != a + 2:
  list_start.append(0)
def koef():
  global a; global list_start
  yield list_start
  for i in range(0, a + 2):
    list_start[a + 1 - i] += list_start[a - i]

while n != a + 1:
  for element in koef():
    element = str(element)
    element = re.sub(r"[\D]", " ", element)
    element = element.replace(" 0 ", "")
    print(element)
    n += 1

print("\033[37m")
```

6. Замість функції з п. 4 створіть генератор, який буде визначати послідовність книжкових сторінок для кожного зошита. Ці рядки можна потім вивести або у циклі, або за допомогою функції next(). Функція-декоратор має працювати з цим генератором так само, як і зі звичайною функцією.

```
print("Enter the format(only 16, 24 or 32):")
while True:
  print_format = input()
  try:
    print_format = int(print_format)
    if print_format != 16 and print_format != 24 and print_format != 32:
      print("\033[31mincorrect format\033[37m")
      continue
  except ValueError:
    print("\033[31mincorrect number\033[37m")
  else:
    break

print("Enter the number of pages(only less than 1280):")
while True:
  page = input()
  try:
    page = int(page)
    if page < 1 or page > 1280:
      print("\033[31mincorrect number of pages\033[37m")
      continue
    elif page%print_format != 0:
      print("\033[31mincorrect number of pages\nthis number must divide on choosen format\033[37m")
      continue
  except ValueError:
    print("\033[31mincorrect number\033[37m")
  else:
    break

print("Enter the output format(1 for lists/ 0 for list):")
while True:
  output_flag = input()
  try:
    output_flag = int(output_flag)
    if output_flag != 1 and output_flag != 0:
      print("\033[31menter 0 or 1\033[37m")
      continue
  except ValueError:
    print("\033[31mincorrect number\033[37m")
  else:
    break

print("\033[32m")
pencilbook = int(page / print_format)
list_for_book = list(range(1, page + 1))
if output_flag == 0:
  value = True
else:
  value = False

def zero_or_one(active=True):
  def wrap(func):
    def wrapper(*args, **kwargs):
      if active:
        final_list = []; n = 0
        pages = func(*args, **kwargs) 
        while n != pencilbook:
          final_list.append(next(pages))
          n += 1
        print(final_list)
      else:
        pages = func(*args, **kwargs)
        n = 0
        while n != pencilbook:
          print(next(pages))
          n += 1
    return wrapper
  return wrap

@zero_or_one(active=value)
def printing(pencilbook, list_for_book):
  for number in range(0, pencilbook):
    adding_list = []
    part = list_for_book[number*print_format + 0 : (number+1)*print_format]
    while len(part) != 0:
      adding_list.append(part[-1])
      adding_list.append(part[0])
      adding_list.append(part[1])
      adding_list.append(part[-2])
      part.remove(part[-1])
      part.remove(part[0])
      part.remove(part[0])
      part.remove(part[-1])
    yield adding_list

printing(pencilbook, list_for_book)
print("\033[37m")
```
